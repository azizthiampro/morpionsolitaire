```python
import matplotlib.pyplot as plt

def plot_grid(grid_size):
    fig, ax = plt.subplots()
    ax.set_aspect('equal')
    ax.set_xticks(range(1,grid_size+1))
    ax.set_yticks(range(1,grid_size+1))
    ax.grid(True, which='both', linestyle='-', linewidth=1)

    plt.show()

# Taille de la grille
grid_size = 21

# Afficher la grille avec lignes horizontales et verticales
plot_grid(grid_size)
```

```python
import math
def get_maltese_cross_branch(width):
    points = []
    for angle in range(0, 360, 45):
        x = width * math.cos(math.radians(angle))
        y = width * math.sin(math.radians(angle))
        points.append((x, y))
    return points

def draw_maltese_cross(ax, center, width):
    branch_points = get_maltese_cross_branch(width)

    # Définir le point central de la croix
    center_x, center_y = center

    # Dessiner chaque branche
    for branch in branch_points:
        x = [center_x + point[0] for point in branch]
        y = [center_y + point[1] for point in branch]
        ax.plot(x, y, color='black', linewidth=1)

# Exemple d'utilisation
grid_size = 21
center = (grid_size / 2, grid_size / 2)
width = 4

fig, ax = plt.subplots()
ax.set_aspect('equal')
ax.set_xticks(range(1,grid_size+1))
ax.set_yticks(range(1,grid_size+1))
ax.grid(True, which='both', linestyle='-', linewidth=1)

draw_maltese_cross(ax, center, width)

plt.show()
```
